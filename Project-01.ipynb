```python
savings_analysis = df.groupby('savings_goal_met')['monthly_income'].mean()
print(f"Average income of those who met goals: ${savings_analysis[1]:.2f}")
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='financial_stress_level', y='discretionary_spending', data=df)
plt.title('Discretionary Spending by Stress Level')
plt.show()
```

```python
df['buffer'] = df['monthly_income'] - (df['essential_spending'] + df['discretionary_spending'] + df['loan_payment'])
df['buffer']
```

```python
red_flags = df[(df['fraud_flag'] == 1) & (df['credit_score'] < 600)]
print(f"Number of high-risk accounts: {len(red_flags)}")
```

```python
# 1. Correlation Heatmap
plt.figure(figsize=(12, 8))
corr = df.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=False, cmap='RdBu', center=0)
plt.title('Feature Correlation Matrix')
plt.savefig('correlation_heatmap.png')
```

```python
# 2. Stress vs. Savings Goal
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='financial_stress_level', hue='savings_goal_met', palette='magma')
plt.title('Impact of Stress on Meeting Savings Goals')
plt.savefig('stress_vs_savings.png')
```

```python
# 3. Income vs. Savings Rate
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='monthly_income', y='savings_rate', hue='income_type', alpha=0.7)
plt.title('Income vs. Savings Rate by Employment Type')
plt.savefig('income_vs_savings_rate.png')
```

```python
# 4. Spending Breakdown
spending_summary = df.groupby('category')[['essential_spending', 'discretionary_spending']].mean()
spending_summary.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Average Essential vs. Discretionary Spending')
plt.ylabel('Amount ($)')
plt.savefig('spending_by_category.png')
```
